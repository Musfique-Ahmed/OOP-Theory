````python
def is_prime(number):
   if number <= 1:
       return False
   for i in range(2, number):
       if number % i == 0:
           return False
   return True
def prime_printer(limit):
   prev = 2
   for i in range(3, limit):
       if is_prime(i):
           if i-prev == 2:
               pair = (prev , i)
               yield pair
           prev = i


a = prime_printer(1000)
for i in a:
   print(i)
````

### Write a decorator called time_measure that measure the running time of a function


````python
import time
import math

def prime_counter_slow(limit):
  count = 0
  for i in range(2, limit):
    is_prime = True
    for j in range(2, i):
      if i % j == 0:
        is_prime = False
        break
    if is_prime:
      count += 1
  return count


def prime_counter_fast(limit):
  count = 0
  for i in range(2, limit):
    is_prime = True
    for j in range(2, int(math.sqrt(i)+1)):
      if i % j == 0:
        is_prime = False
        break
    if is_prime:
      count += 1
  return count




def time_measure(func):
  def wrapper(limit):
    t1 = time.time()
    print(func(limit))
    t2 = time.time()
    print(f'Time: {t2-t1}')
  return wrapper

@time_measure
def prime_counter_slow(limit):
  count = 0
  for i in range(2, limit):
    is_prime = True
    for j in range(2, i):
      if i % j == 0:
        is_prime = False
        break
    if is_prime:
      count += 1
  return count


@time_measure
def prime_counter_fast(limit):
  count = 0
  for i in range(2, limit):
    is_prime = True
    for j in range(2, int(math.sqrt(i)+1)):
      if i % j == 0:
        is_prime = False
        break
    if is_prime:
      count += 1
  return count


prime_counter_slow(10000)
prime_counter_fast(10000) 
````

# Decorator 242 
## 2(a) 

Youâ€™ve bought a new earphone and want to give it to your 
grandmother, so you want to ensure she is not exposed to unsafe 
sound levels. Each time the earphone attempts to play an audio clip, it 
first checks the highest volume level in that clip. If the maximum 
volume is within the allowed safe limit, the clip is played normally. 
However, if the volume exceeds the predefined safe threshold, 
playback is immediately stopped and a custom 
MaxVolumeExceededError is raised to prevent loud sound 
exposure. 


Write a decorator volume_limiter to implement this 
volume-checking mechanism, ensuring that the safety logic is applied 
automatically before the audio is played, without modifying the core 
playback functionality. The code should produce the desired output 
given below. 

````python
class MaxVolumeExceededError(Exception):
    pass

def volume_limiter(func): #decorator function that takes a function as an argument and returns a new function that checks the volume before calling the original function
    def wrapper(self, clip): #self is the instance of the class and clip is the argument of the function
        max_volume = self.check_volume(clip) #check_volume is a method of the class that checks the volume of the clip and returns the maximum volume
        if max_volume <= self.desired_level:
            return func(self, clip)
        else:
            raise MaxVolumeExceededError(f"{max_volume} exceeded the limit: {self.desired_level}")
    return wrapper

class Earphone:
    def __init__(self):
        self.desired_level = 85
        
    def check_volume(self, clip):
        return max(clip.volumes)
        
    @volume_limiter #decorator
    def play(self, clip):
        print("Playing clip with volumes:", clip.volumes)

class Clip:
    def __init__(self, volumes):
        self.volumes = volumes

bose = Earphone()
clip1 = Clip([12, 34, 56, 63, 85])
clip2 = Clip([12, 34, 56, 63, 89])

try:
    bose.play(clip1)
    bose.play(clip2)
except MaxVolumeExceededError as e:
    print(e)
````

````python
def dec1(func):
    def wrapper():
        print("1")
        func()
    return wrapper

def dec2(func):
    count = 1
    def wrapper():
        nonlocal count
        if count > 0:
            func()
            count -= 1
    return wrapper

def dec3(func):
    def wrapper():
        print("3")
        func()
    return wrapper


@dec2
@dec1
@dec3
def show():
    print("Hi")


show()
show()
````

# Tracing Output Explanation

### Decorator Structure
The decorators are applied from bottom to top (inner to outer).
```python
@dec2
@dec1
@dec3
def show(): ...
```
This is equivalent to: `show = dec2(dec1(dec3(show)))`

### Execution Flow

#### 1. First Call: `show()`
*   **Enters `dec2` wrapper**: The variable `count` is initialized to `1` from the closure.
    *   Checks `if count (1) > 0`: **True**.
    *   Calls the inner function (`dec1` wrapper).
*   **Enters `dec1` wrapper**:
    *   Prints **`"1"`**.
    *   Calls the inner function (`dec3` wrapper).
*   **Enters `dec3` wrapper**:
    *   Prints **`"3"`**.
    *   Calls the original `show` function.
*   **Original `show` function**:
    *   Prints **`"Hi"`**.
*   **Returns to `dec2` wrapper**:
    *   Decrements `count`. `count` becomes `0`.

**Output for 1st Call:**
```
1
3
Hi
```

#### 2. Second Call: `show()`
*   **Enters `dec2` wrapper**: The variable `count` is now `0`.
    *   Checks `if count (0) > 0`: **False**.
    *   The body is skipped. The inner functions are never called.

**Output for 2nd Call:**
*(Nothing involves printing)*
