```python
class NegativeWithdrawalError(Exception):
   pass
def withdraw():
   try:
       amount = int(input('Enter withdrawal amount: '))
       if amount < 0:
           raise NegativeWithdrawalError
   except NegativeWithdrawalError:
       print("Please enter withdrawal amount greater than 0")
   else:
       return amount
transactionHistory = []
transactionHistory.append(withdraw())
print(transactionHistory)
```

```python
class EmptyListError(Exception):
    pass
class ElementNotANumberError(Exception):
    pass
class ThresholdNotanIntError(Exception):
    pass
class ElementOutOfRangeError(Exception):
    pass


def processSensorData(temp_list, threshold):
  filtered_list = []
  try:
      if len(temp_list) == 0:
        raise EmptyListError("Empty list")
      if isinstance(threshold, int) == False:
        raise ThresholdNotanIntError("Threshold not int")
      for i in temp_list:
        if isinstance(i, (int, float)) == False:
          raise ElementNotANumberError(f"{i} Element Not A Number Error")
        if i > 100 or i <-100:
          raise ElementOutOfRangeError(f"{i} Elementout of range Error")
        if i > threshold:
          filtered_list.append(i)
  except Exception as e:
    print(e)
    return "sakib"
  else:
    return filtered_list


a = [1,20,3,4,5,"6",7000]


print(processSensorData(a, 4))
```
