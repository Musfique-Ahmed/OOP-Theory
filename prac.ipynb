```python
class Point:
    def __init__(self, xCor = 0, yCor=0):
        self.x = xCor
        self.y = yCor
    def show(self):
        print(f"({self.x},{self.y})")   
    def increaseX(self,val):
        self.x+=val
    def increaseY(self,val):
        self.y+=val
p1 = Point(50,50)
p1.show()

p1.increaseX(5)
p1.increaseY(-5)
p1.show()
```

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

 # Official representation (for developers/debugging)
    def __repr__(self):
        return f"Point(x={self.x}, y={self.y})"

 # User-friendly representation (for end-users)
    def __str__(self):
        return f"({self.x}, {self.y})"

p = Point(10, 20)

 #  print() and str() call __str__()
print(p) # Output: (10, 20)
print(str(p)) # Output: (10, 20)

# 2. Interactive console and repr() call __repr__()
print(repr(p)) # Output: Point(x=10, y=20)
```

```python
class Account:
    def __init__(self, initial_balance = 500):
# "Private" attribute (name mangling)
        self.__balance = initial_balance
# "Protected" attribute (convention)
        self._owner_name = "Anonymous"
# Public method to access the balance
    def get_balance(self):
        return self.__balance

    # "Private" method (name mangling)
    def __update_balance(self, amount):
        self.__balance += amount
        print("Balance updated internally.")

    def deposit(self, amount):
        if amount > 0:
            # We call the private method internally
            self.__update_balance(amount)

acc = Account(500)
acc.deposit(100)
print(f"Current Balance: {acc.get_balance()}")

# acc.__update_balance(200)  # This will raise an AttributeError

# External attempts (Discouraged/Failed/Difficult)
print(f"Owner Name (Access discouraged): {acc._owner_name}")

# Accessing mangled attribute (Technically possible, but bad practice)
# print(f"Mangled Balance: {acc._Account__balance}")
```

```python
class Time:
    def __init__(self,h=0,m=0):
        self.hour = h
        self.min=m
    def __add__(self,other):
        return Time(self.hour+other.hour,self.min+other.min)
    def __str__(self):
        return f"{self.hour} hours {self.min} mins"
t1 = Time(10,30)
t2 = Time(5,10)
print(t1+t2)
```

```python
class Sword:
   def __init__(self, name, weight):
       self.name = name
       self.weight = weight
   def __add__(self, other):
       combined_name = self.name + '-' + other.name
       combined_weight = self.weight + other.weight
       return Sword(combined_name, combined_weight)
   def __gt__(self, other):
       return self.weight > other.weight
   def __str__(self):
       return f"{self.name} (Weight: {self.weight} kg)"
sword1 = Sword("Dragon Fang", 3.5)
sword2 = Sword("Phoenix Claw", 4.2)
legendary_sword = Sword("Excalibur", 8.0)
fusion_sword = sword1 + sword2
print(f"Combined Sword: {fusion_sword}")
if fusion_sword > legendary_sword:
   print(f"{fusion_sword.name} is heavier than {legendary_sword.name}.")
else:
   print(f"{fusion_sword.name} is lighter than {legendary_sword.name}.")
```

```python
from copy import copy, deepcopy
class Student:
   def __init__(self, name, courses):
       self.name = name
       self.courses = courses

# DEEP COPY FUNCTION
def copy_student(obj):
   lst = []
   for c in obj.courses:
       lst.append(c)
   return Student(obj.name, lst)

#shallow copy function
def copy_new_student(obj):
    return obj


kt = Student("Kotha", ["OOP", "OOP lab"])
kt_us = copy_student(kt) #deep copy
kt_py = copy(kt) #shallow copy
kt_deep = deepcopy(kt) #deep copy
kt_new = copy_new_student(kt)
kt.courses.append("ENG")


print(kt.courses)
print(kt_us.courses)
print(kt_py.courses)
print(kt_deep.courses)
print(kt_new.courses)
```
