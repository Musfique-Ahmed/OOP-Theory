```python
class Point:
    def __init__(self, xCor = 0, yCor=0):
        self.x = xCor
        self.y = yCor
    def show(self):
        print(f"({self.x},{self.y})")   
    def increaseX(self,val):
        self.x+=val
    def increaseY(self,val):
        self.y+=val
p1 = Point(50,50)
p1.show()

p1.increaseX(5)
p1.increaseY(-5)
p1.show()
```

```python
class Rectangle:
    def __init__(self,w,h):
        self.width=w
        self.height=h
    def getArea(self):
        return self.height*self.width
r = Rectangle(4,12)
print(r.getArea())
```

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

 # Official representation (for developers/debugging)
    def __repr__(self):
        return f"Point(x={self.x}, y={self.y})"

 # User-friendly representation (for end-users)
    def __str__(self):
        return f"({self.x}, {self.y})"

p = Point(10, 20)

 #  print() and str() call __str__()
print(p) # Output: (10, 20)
print(str(p)) # Output: (10, 20)

# 2. Interactive console and repr() call __repr__()
print(repr(p)) # Output: Point(x=10, y=20)
```

```python
class Account:
    def __init__(self, initial_balance = 500):
# "Private" attribute (name mangling)
        self.__balance = initial_balance
# "Protected" attribute (convention)
        self._owner_name = "Anonymous"
# Public method to access the balance
    def get_balance(self):
        return self.__balance

    # "Private" method (name mangling)
    def __update_balance(self, amount):
        self.__balance += amount
        print("Balance updated internally.")

    def deposit(self, amount):
        if amount > 0:
            # We call the private method internally
            self.__update_balance(amount)

acc = Account(500)
acc.deposit(100)
print(f"Current Balance: {acc.get_balance()}")

# acc.__update_balance(200)  # This will raise an AttributeError

# External attempts (Discouraged/Failed/Difficult)
print(f"Owner Name (Access discouraged): {acc._owner_name}")

# Accessing mangled attribute (Technically possible, but bad practice)
# print(f"Mangled Balance: {acc._Account__balance}")
```

```python
class Course:
    def __init__(self, course_name, credits):
        self.course_name = course_name
        self.credits = credits

    def display_info(self):
        print(f"Course Name: {self.course_name}, Credits: {self.credits}")


bangla = Course("Bangla", 3)
bangla.display_info()

english = Course("English", 3)
english.display_info()
```

```python
class Time:
    def __init__(self,h=0,m=0):
        self.hour = h
        self.min=m
    def __add__(self,other):
        return Time(self.hour+other.hour,self.min+other.min)
    def __str__(self):
        return f"{self.hour} hours {self.min} mins"
t1 = Time(10,30)
t2 = Time(5,10)
print(t1+t2)
```
