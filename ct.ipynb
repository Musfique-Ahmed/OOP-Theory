```python
class Point3D:
  def __init__(self, x, y, z):
   self.x=x
   self.y=y
   self.z=z
  def copy(self):
   return Point3D(self.x, self.y, self.z)
  def __str__(self):
   return f"({self.x},{self.y}, {self.z})"
  def scale(self, n):
   self.x *= self.x
   self.y *= self.y
   self.z *= self.z



p1 = Point3D(1, 2, 3)
p2 = p1
p3 = p2.copy()
p1.scale(2)
p2.scale(2)
p3.scale(2)
print(p1, p2, p3)
```
